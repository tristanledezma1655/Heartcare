```python
def calcular_metricas_manuales(y_true, y_pred, nombre_modelo="Modelo"):
    # Convertimos a listas de Python por si vienen de NumPy o Pandas
    y_true = list(y_true)
    y_pred = list(y_pred)

    tp, tn, fp, fn = 0, 0, 0, 0

    # 1. Contamos uno a uno los aciertos y errores (Puro CC)
    for i in range(len(y_true)):
        if y_true[i] == 1 and y_pred[i] == 1:
            tp += 1
        elif y_true[i] == 0 and y_pred[i] == 0:
            tn += 1
        elif y_true[i] == 0 and y_pred[i] == 1:
            fp += 1
        elif y_true[i] == 1 and y_pred[i] == 0:
            fn += 1

    # 2. Calculamos las fórmulas matemáticas (evitando división por cero)
    accuracy = (tp + tn) / len(y_true) if len(y_true) > 0 else 0
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    # 3. Imprimimos un reporte hermoso para la presentación
    print(f"==========================================")
    print(f"MÉTRICAS MANUALES PARA: {nombre_modelo.upper()}")
    print(f"==========================================")
    print(f"Verdaderos Positivos (TP): {tp}")
    print(f"Falsos Positivos (FP):     {fp}")
    print(f"Verdaderos Negativos (TN): {tn}")
    print(f"Falsos Negativos (FN):     {fn}")
    print(f"------------------------------------------")
    print(f"Accuracy:  {accuracy:.4f} ({accuracy*100:.1f}%)")
    print(f"Precision: {precision:.4f} ({precision*100:.1f}%)")
    print(f"Recall:    {recall:.4f} ({recall*100:.1f}%)")
    print(f"F1-Score:  {f1_score:.4f} ({f1_score*100:.1f}%)")
    print(f"==========================================\n")
```

```python
import time
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

# --- 1. PREPARACIÓN DE TODOS LOS ARRAYS EN NUMPY PURO ---
X_train_cc = x_cc_train_proc.toarray() if hasattr(x_cc_train_proc, "toarray") else np.asarray(x_cc_train_proc)
y_train_cc = np.asarray(y_cc_train)
X_test_cc = x_cc_test_proc.toarray() if hasattr(x_cc_test_proc, "toarray") else np.asarray(x_cc_test_proc)

# Dataset Completo (Asegúrate de cambiar estos nombres por tus variables reales de 41k)
X_train_big = x_sk_train_proc.toarray() if hasattr(x_sk_train_proc, "toarray") else np.asarray(x_sk_train_proc)
y_train_big = np.asarray(y_train_sk)
X_test_big = x_sk_test_proc.toarray() if hasattr(x_sk_test_proc, "toarray") else np.asarray(x_sk_test_proc)

# --- 2. DEFINICIÓN DE TAMAÑOS SEPARADOS ---
tamanos_manual = [50, 100, 150, 200, 250, len(X_train_cc)]
tamanos_sklearn = [1000, 5000, 10000, 20000, 30000, len(X_train_big)]

tiempos_lr_manual = []
tiempos_knn_manual = []
tiempos_lr_sklearn = []
tiempos_knn_sklearn = []

# --- 3. EXPERIMENTO DE TIEMPOS ---
# Medición para Modelos Manuales (Muestra corta)
for n in tamanos_manual:
    X_sub, y_sub = X_train_cc[:n], y_train_cc[:n]

    # LR Manual
    lr_m = RegresionLogisticaPura(lr=0.01, epochs=100)
    inicio = time.time()
    lr_m.fit(X_sub, y_sub)
    tiempos_lr_manual.append(time.time() - inicio)

    # k-NN Manual
    knn_m = KNNPuro(k=5)
    knn_m.fit(X_sub, y_sub)
    inicio = time.time()
    _ = knn_m.predict(X_test_cc)
    tiempos_knn_manual.append(time.time() - inicio)

# Medición para Sklearn (Dataset Grande y con MÁXIMA OPTIMIZACIÓN)
for n in tamanos_sklearn:
    X_sub, y_sub = X_train_big[:n], y_train_big[:n]

    # LR Sklearn
    lr_sk = LogisticRegression(max_iter=100, solver='saga', random_state=42)
    inicio = time.time()
    lr_sk.fit(X_sub, y_sub)
    tiempos_lr_sklearn.append(time.time() - inicio)

    # k-NN Sklearn ( Quitamos 'brute' para activar KD-Tree/Ball-Tree automáticos)
    knn_sk = KNeighborsClassifier(n_neighbors=5, algorithm='auto')
    knn_sk.fit(X_sub, y_sub)
    inicio = time.time()
    _ = knn_sk.predict(X_test_big)
    tiempos_knn_sklearn.append(time.time() - inicio)


# --- 4. GRAFICACIÓN EN PANEL DE 2x2 ---
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Análisis Asintótico (Manual vs Máxima Optimización Sklearn)', fontsize=16, fontweight='bold')

# Ejes X continuos para las curvas teóricas suavizadas
x_linea_manual = np.linspace(min(tamanos_manual), max(tamanos_manual), 100)
x_linea_sklearn = np.linspace(min(tamanos_sklearn), max(tamanos_sklearn), 100)

# Gráfica 1: LR Manual (Lineal)
axs[0, 0].plot(tamanos_manual, tiempos_lr_manual, 'o-', color='firebrick', label='Observado')
p1 = np.polyfit(tamanos_manual, tiempos_lr_manual, 1)
axs[0, 0].plot(x_linea_manual, np.polyval(p1, x_linea_manual), '--', color='darkred', label='O(E * N * M) Teórico')
axs[0, 0].set_title('1. LR Manual (Fit - Muestreado)')
axs[0, 0].set_xlabel('Tamaño de Muestra (N)')
axs[0, 0].set_ylabel('Tiempo (Segundos)')
axs[0, 0].grid(True, linestyle=':')
axs[0, 0].legend()

# Gráfica 2: k-NN Manual (Lineal respecto a N)
axs[0, 1].plot(tamanos_manual, tiempos_knn_manual, 'o-', color='darkorange', label='Observado')
p2 = np.polyfit(tamanos_manual, tiempos_knn_manual, 1)
axs[0, 1].plot(x_linea_manual, np.polyval(p2, x_linea_manual), '--', color='chocolate', label='O(T * N * M) Teórico')
axs[0, 1].set_title('2. k-NN Manual (Predict - Muestreado)')
axs[0, 1].set_xlabel('Tamaño de Train (N)')
axs[0, 1].set_ylabel('Tiempo en Test (Segundos)')
axs[0, 1].grid(True, linestyle=':')
axs[0, 1].legend()

# Gráfica 3: LR Sklearn (Lineal de pendiente ultra-baja)
axs[1, 0].plot(tamanos_sklearn, tiempos_lr_sklearn, 's-', color='navy', label='Observado')
p3 = np.polyfit(tamanos_sklearn, tiempos_lr_sklearn, 1)
axs[1, 0].plot(x_linea_sklearn, np.polyval(p3, x_linea_sklearn), '--', color='dodgerblue', label='O(N) optimizado Vectorizado')
axs[1, 0].set_title('3. LR Sklearn (Fit - Dataset completo)')
axs[1, 0].set_xlabel('Tamaño de Dataset completo (N)')
axs[1, 0].set_ylabel('Tiempo (Segundos)')
axs[1, 0].grid(True, linestyle=':')
axs[1, 0].legend()

# Gráfica 4: k-NN Sklearn ( LOGARÍTMICA REAL)
axs[1, 1].plot(tamanos_sklearn, tiempos_knn_sklearn, 's-', color='teal', label='Observado')
# Ajustamos una función logarítmica real sobre los puntos empíricos medidos
coef_log = np.polyfit(np.log(tamanos_sklearn), tiempos_knn_sklearn, 1)
y_log_teorico = coef_log[0] * np.log(x_linea_sklearn) + coef_log[1]
axs[1, 1].plot(x_linea_sklearn, y_log_teorico, '--', color='lightseagreen', label='O(T * log N) optimizado con Árboles Indexados')
axs[1, 1].set_title('4. k-NN Sklearn (Predict - Dataset completo)')
axs[1, 1].set_xlabel('Tamaño de Train completo (N)')
axs[1, 1].set_ylabel('Tiempo en Test (Segundos)')
axs[1, 1].grid(True, linestyle=':')
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```

```python
%%writefile /content/Heartcare/Regresion_Logistica.py

import math
import numpy as np

class RegresionLogisticaPura:
    def __init__(self, lr=0.01, epochs=100):
        self.lr = lr
        self.epochs = epochs
        self.weights = None
        self.bias = 0.0

    def fit(self, X, y):
        # X e y deben ser listas o arrays simples de NumPy
        X = np.array(X)
        y = np.array(y)

        n_samples = len(X)
        n_features = len(X[0])

        # Inicializamos los pesos en cero para cada variable
        self.weights = [0.0] * n_features
        self.bias = 0.0

        # Bucle de épocas (iteraciones globales)
        for _ in range(self.epochs):
            # Pasamos por cada uno de los 360 datos individualmente
            for i in range(n_samples):
                # 1. Calcular Z (Xw + b) usando un bucle for para el producto punto
                z = self.bias
                for j in range(n_features):
                    z += X[i][j] * self.weights[j]

                # 2. Aplicar la función Sigmoide manual
                # Limitamos z entre -50 y 50 para evitar errores de desbordamiento en math.exp
                z = max(-50, min(50, z))
                y_pred = 1.0 / (1.0 + math.exp(-z))

                # 3. Calcular el error del dato actual
                error = y_pred - y[i]

                # 4. Actualizar los pesos uno por uno (Gradiente Descendiente)
                for j in range(n_features):
                    self.weights[j] -= self.lr * error * X[i][j]
                self.bias -= self.lr * error

    def predict(self, X):
        predicciones = []
        for i in range(len(X)):
            z = self.bias
            for j in range(len(X[0])):
                z += X[i][j] * self.weights[j]

            z = max(-50, min(50, z))
            y_pred = 1.0 / (1.0 + math.exp(-z))

            # Clasificación binaria con umbral de 0.5
            if y_pred >= 0.5:
                predicciones.append(1)
            else:
                predicciones.append(0)
        return predicciones
```

```python
%%writefile /content/Heartcare/Knn.py

class KNNPuro:
    def __init__(self, k=5):
        self.k = k
        self.X_train = []
        self.y_train = []

    def fit(self, X, y):
        # En k-NN manual solo guardamos la muestra en memoria
        self.X_train = X
        self.y_train = y

    def predict(self, X_test):
        predicciones = []

        # Para cada paciente en el conjunto de prueba...
        for x_test_individual in X_test:
            distancias = []

            # ...calculamos la distancia contra cada uno de los 360 pacientes de entrenamiento
            for idx, x_train_individual in enumerate(self.X_train):
                suma_cuadrados = 0.0

                # Bucle para recorrer cada variable/columna una por una
                for j in range(len(x_test_individual)):
                    suma_cuadrados += (x_test_individual[j] - x_train_individual[j]) ** 2

                distancia_euclidiana = math.sqrt(suma_cuadrados)
                # Guardamos la distancia junto con la etiqueta (sano/enfermo) de ese vecino
                distancias.append((distancia_euclidiana, self.y_train[idx]))

            # Ordenamos las distancias de menor a mayor (Fuerza Bruta pura)
            distancias.sort(key=lambda x: x[0])

            # Tomamos los K vecinos más cercanos
            k_vecinos = distancias[:self.k]

            # Contamos los votos (cuántos enfermos y cuántos sanos)
            votos_enfermos = sum(1 for d, etiqueta in k_vecinos if etiqueta == 1)
            votos_sanos = self.k - votos_enfermos

            # El voto mayoritario gana
            if votos_enfermos > votos_sanos:
                predicciones.append(1)
            else:
                predicciones.append(0)

        return predicciones
```
